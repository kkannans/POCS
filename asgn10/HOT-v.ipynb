```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
def create_forest(L, p):
    forest = np.array([[1 if np.random.random() < p else 0 for _ in range(L)] for _ in range(L)])
    
    return forest

def label_forest(forest):
    #filter largest cluster
    structure = [[0,1,0],[1,1,1],[0,1,0]] #define connection
    label_forest, nb_labels = ndimage.label(forest,structure) #label clusters
    sizes = ndimage.sum(forest, label_forest, range(nb_labels + 1))
    return label_forest, sizes
create_forest(10, 0.5)
```

```python
def p_matrix(L, l):
    p_mat = np.zeros((L,L))
    for i in range(L):
        for j in range(L):
            p_mat[i][j] = np.exp(-i/l)*np.exp(-j/l)
    return p_mat / np.sum(p_mat)
p_matrix(4, 1) 
```

```python
def forest_labeled_with_size(forest, sizes):
    size_forest = forest.copy()
    for i, size in enumerate(sizes[1:]):
        size_forest[size_forest == i+1] = size
    return size_forest
```

```python
forest = create_forest(10, 0.5)
forest, sizes = label_forest(forest)
size_forest = forest_labeled_with_size(forest, sizes)
print(forest, '\n', sizes, '\n' , size_forest)
```

```python
import copy
def plant_one_tree(forest, empty, p_mat, D):
    least_destruction = np.inf
    least_destruction_tree = None

    sim_empty = copy.deepcopy(empty)
    
    # Simulate D tree plants, see what the yield would be
    # Maximize yield, minimize P*S
    for d in range(D):
        if len(sim_empty) == 0: 
            continue
        sim_forest = forest.copy()
        tree_loc_idx = np.random.choice(range(len(sim_empty)))
        tree_loc = sim_empty[tree_loc_idx]
        sim_empty.remove(tree_loc)
        
        # plant da tree
        sim_forest[tree_loc] = 1
        # label da forest
        sim_forest_labeled, sizes = label_forest(sim_forest)
        # count da yield
        size_forest = forest_labeled_with_size(sim_forest_labeled, sizes)
        
#         print(f'd={d}\n{tree_loc}\n', sim_forest)
        
        # See if it's a betta tree
        destruction = np.sum(p_mat * size_forest)
        if destruction < least_destruction:
            least_destruction_tree = tree_loc
            least_destruction = destruction
            
    forest[least_destruction_tree] = 1
    return forest, least_destruction_tree, least_destruction
```

```python
def run_one_sim(L, D, rho):
    # Start with empty forest
    print(f"L ={L}, D = {D}")
    forest = np.zeros((L, L))
    empty = [(i,j) for i in range(L) for j in range(L)]
    planted = []
    
    trees_planted = 0
    
    # Probability matrix over forest
    p_mat = p_matrix(L, rho)
    
    yieldz = []
    max_yield_forest = None
    max_yield = 0
    
    # Keep planting trees until ? yield starts falling? 
    while trees_planted < L**2:
        if trees_planted % 200 == 0:
            print('trees planted: ', trees_planted)
            
        # Plant a tree
        forest, tree_loc, destruction = plant_one_tree(forest, empty, p_mat, D)
        trees_planted += 1
        
        # Compute yield
        yeeld = trees_planted - destruction
        if yeeld > max_yield:
            max_yield = yeeld
            max_yield_forest = forest.copy()
        
        # Add to yields
        yieldz.append((trees_planted, yeeld))
        planted.append(tree_loc)
        empty.remove(tree_loc)
    
    return yieldz, max_yield_forest, max_yield
```

```python
def do_a_run(L, D):
    plt.rcParams['figure.figsize'] = (10,5)
    fig, [maxyield_ax, im_ax] = plt.subplots(1, 2)
    y, max_forest, max_yield = run_one_sim(L, D, L/10)
    maxyield_ax.scatter(*zip(*y))
    maxyield_ax.set_xlabel('Num trees added')
    maxyield_ax.set_ylabel('Avg Yield')
    maxyield_ax.set_title(f'L={L}, D={D} Yield vs num trees added')
    img = im_ax.imshow(max_forest)
    im_ax.set_title('Max yield forest')
    fig.colorbar(img)
    fig_name = f"L_{L}_D_{D}_yield_over_trees_and_max_forest.png"
    plt.savefig(fig_name)
```

```python
def zipf_plot(L, D, rho, ax):
    print(f'Running size {L} with D={D}')
    y, max_forest, max_yield = run_one_sim(L, D, rho)
    labeled_forest, sizes = label_forest(max_forest)
    sorted_sizes = sorted(sizes, reverse=True)
    points = [(np.log10(i+1), np.log10(size)) for i,size in enumerate(sorted_sizes)]
    ax.scatter(*zip(*points))
    ax.set_xlabel('log10(rank)')
    ax.set_ylabel('log10(tree component size)')
    ax.set_title(f'L={L}, D={D} zipf')
```

```python
plt.rcParams['figure.figsize'] = (10,6)
D_vals = [1,2,3,L, L*L]
fig, ax = plt.subplots(1, len(D_vals))
L = 32
for j, D in enumerate(D_vals):
    zipf_plot(L, D, ax[j])
plt.tight_layout()
plt.show()
plt.savefig("zipf_plots_tree_component_size_for_D_vals.png")
```

```python
def zipf_plot(L, D, rho, ax):
    print(f'Running size {L} with D={D}')
    y, max_forest, max_yield = run_one_sim(L, D, rho)
    labeled_forest, sizes = label_forest(max_forest)
    sorted_sizes = sorted(sizes, reverse=True)
    points = [(np.log10(i+1), np.log10(size)) for i,size in enumerate(sorted_sizes)]
    ax.scatter(*zip(*points))
    ax.set_xlabel('log10(rank)')
    ax.set_ylabel('log10(tree component size)')
    ax.set_title(f'L={L}, D={D} zipf')
```

```python
# Zipf for D = L^2 for rho = [0.1, .. 0.9]
rhos = [i/10 for i in range(1,10)]
plt.rcParams['figure.figsize'] = (10,10)
L = 32
D = L**2
fig, ax = plt.subplots(3, 3)
print(len(rhos))
for idx, rho in enumerate(rhos):
    print(f"rho = {rho}")
    i = idx//3
    j = idx%3
    zipf_plot(L, D, rho, ax[i,j])
plt.tight_layout()
plt.show()
figure_name = "zipf_plots_L_32_D_L^2_rhos.png"
plt.savefig(figure_name)
```
