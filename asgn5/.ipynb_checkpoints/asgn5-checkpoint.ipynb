```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import glob
from collections import Counter
from collections import OrderedDict 
```

# Plot time series for the rank of these baby names in the US

```python
baby_names_files = glob.glob("data/*")
required_names = ["Shirley", "Desmond", "Madison", "Aiden", "Daniel"]
data_names = {"boys":{}, "girls":{}} 
# boys first
for file in baby_names_files:
    f = open(file,'r')
    if "boys" in file:
        key = file.replace("data/names-boys", "")
        key = int(key.replace(".txt", "")) # should be a year
    elif "girls" in file:
        key = file.replace("data/names-girls", "")
        key = int(key.replace(".txt", "")) # should be a year
    else:
        continue
    data = f.read().split() # read
    records = np.array([s.split(",") for s in data]) # get array of list of str
    names = records[:,0]            
    freqs = records[:,2]
    freqs = [int(freq) for freq in freqs]
    # dict of names to frequencies
    names_to_freqs = {x:y for x,y in zip(names, freqs)}
    # sort by reverse order
    names_to_freqs = {k:v for k,v in sorted(names_to_freqs.items(), key = lambda x: x[1], reverse = True)}
    # rank from freqs
    ranks = np.arange(1, len(freqs))
    # name to rank
    names_to_rank = {x:y for x,y in zip(list(names_to_freqs.keys()),ranks)}
    if "boys" in file:
        data_names["boys"][key] = names_to_rank
    elif "girls" in file:
        data_names["girls"][key] = names_to_rank
    else:
        continue
    f.close()
```

```python
# Boys Analysis
def boys_analysis():
    # get required name ranks by yr(time-series) in boys
    # result = {Shirley: [tuple(yr1, rank),tuple(yr2, rank)] , ...}
    result = {name:[] for name in required_names}
    yrs = list(data_names["boys"].keys())
    for yr in yrs:
        all_names = list(data_names["boys"][yr].keys())
        for name in required_names:
            if name in all_names:
                result[name].append(tuple([yr, data_names["boys"][yr][name]]))
    # sort by yr
    for name in result.keys():
        result[name] = sorted(result[name], key=lambda x: x[0])
    # required_names = ["Shirley", "Desmond", "Madison", "Aiden", "Daniel"]
    for key,val in result.items():
        yrs = list(zip(*val))[0]
        ranks = list(zip(*val))[1]
        fig, ax = plt.subplots()
        ax.set_xlabel("yrs")
        ax.set_ylabel("rank")
        ax.plot(yrs, ranks, label = key)
        ax.set_title("Time series of Rank for "+key + " under Boys category")
        plt.savefig("boys_category_"+key+"_time_series_rank.png")
        plt.legend()
        plt.show()
```

```python
# Girls Analysis
def girls_analysis():
    # get required name ranks by yr(time-series) in boys
    # result = {Shirley: [tuple(yr1, rank),tuple(yr2, rank)] , ...}
    result = {name:[] for name in required_names}
    yrs = list(data_names["girls"].keys())
    for yr in yrs:
        all_names = list(data_names["girls"][yr].keys())
        for name in required_names:
            if name in all_names:
                result[name].append(tuple([yr, data_names["girls"][yr][name]]))
    # sort by yr
    for name in result.keys():
        result[name] = sorted(result[name], key=lambda x: x[0])
    # required_names = ["Shirley", "Desmond", "Madison", "Aiden", "Daniel"]
    for key,val in result.items():
        yrs = list(zip(*val))[0]
        ranks = list(zip(*val))[1]
        fig, ax = plt.subplots()
        ax.set_xlabel("yrs")
        ax.set_ylabel("rank")
        ax.plot(yrs, ranks, label = key)
        ax.set_title("Time series of Rank for "+key + " under Girls category")
        plt.legend()
        plt.savefig("girls_category_"+key+"_time_series_rank.png")
        plt.show()
```

```python
boys_analysis()
```

```python
girls_analysis()
```

```python
def alpha_0(rank1:int, rank2:int):
    """
    abs(1/r1^alpha - 1/r2^alpha), lim alpha ->0 
    abs(ln(r1/r2))
    """
    assert rank2 > 0
    return abs(np.ln(rank1/rank2))
    
def alpha_inf(rank1:int, rank2:int):
    """
    abs(1/r1^alpha - 1/r2^alpha), lim alpha ->inf
    max(l/r1, 1/r2)
    """
    assert rank1 > 0
    assert rank2 > 0
    return max(1/rank1, 1/rank2) 

def compute_rank_turbulence(baby_names_1:dict, baby_names_2: dict, limits = 0):
    """
    Given baby_names->rank map of system 1, 2
    For each name in 1, find name in 2.
    if not in 2, use assign lowest rank in 2.
    if alpha = 0, use alpha_0
    if alpha = in, use alpha_inf
    store in dict("name": value)
    return dict
    """
    limits = [0]
    rank_turbulence = {}
    baby_names_1 = list(baby_names_1.keys())
    min_rank2 = min(baby_names_2.values())
    for baby_name_1 in baby_names_1:
        if baby_name_1 in baby_names_2.keys():
            rank1 = baby_names_1[baby_name_1]
            rank2 = baby_names_2[baby_name_1]
        else:
            rank2 = min_rank2
        rank_turbulence[baby_name_1] = {
                "alpha_0" : alpha_0(rank1, rank2),
                "alpha_inf": alpha_inf(rank1, rank2)
            } 
    return rank_turbulence
```
